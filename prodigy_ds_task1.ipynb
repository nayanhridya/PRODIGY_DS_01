```python
# Data visualization
def add_value_labels(ax,spacing=10):
    for rect in ax.patches:
        x=rect.get_x()+rect.get_width()/2
        y=rect.get_height()-3

        va='bottom' if y>=0 else 'top'
        label= '()'.format(y)
        y_shift = spacing * (1 if y >= 0 else -1)
        ax.annotate(label,(x,y),xytext=(0,y_shift),textcoords="offset points",ha='center',va=va)
```

DATA VISUALIZATION

Total Population in year 2022

```python
# Select the specific year for which you want to create the bar plot
year = '2022'

# Filter the data for the selected year
df_year = df[['Country Name', year]]

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(df_year['Country Name'], df_year[year], color='skyblue')
plt.title(f"Total Population in {year}")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```

Top 10 countries in the year 2022

```python
# Sort the data by total population (assuming the column name is '2022', adjust if needed)
df_sorted = df.sort_values(by='2022', ascending=False)

# Select the top ten countries
top_ten_countries = df_sorted.head(10)

# Plot the data
plt.figure(figsize=(10, 6))
plt.barh(top_ten_countries['Country Name'], top_ten_countries['2022'], color='skyblue')
plt.title("Top Ten Countries by Total Population (2022)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.gca().invert_yaxis()  # Invert y-axis to display the country with the highest population at the top
plt.tight_layout()
plt.show()
```
